```python
#Q2
def find_dominant_strategies(matrix):
    num_rows = len(matrix)
    num_cols = len(matrix[0])

    p1_dominant = None
    p2_dominant = None

    for r1 in range(num_rows):
        is_r1_dominant = True
        for r2 in range(num_rows):
            if r1 == r2:
                continue
            if not all(
                matrix[r1][c][0] > matrix[r2][c][0] for c in range(num_cols)
            ):
                is_r1_dominant = False
                break
        if is_r1_dominant:
            p1_dominant = r1
            break

    for c1 in range(num_cols):
        is_c1_dominant = True
        for c2 in range(num_cols):
            if c1 == c2:
                continue
            if not all(
                matrix[r][c1][1] > matrix[r][c2][1] for r in range(num_rows)
            ):
                is_c1_dominant = False
                break
        if is_c1_dominant:
            p2_dominant = c1
            break

    if p1_dominant is not None:
        print(f"Player 1 has a strictly dominant strategy: Row {p1_dominant}")
    else:
        print("Player 1 has no strictly dominant strategy.")

    if p2_dominant is not None:
        print(f"Player 2 has a strictly dominant strategy: Column {p2_dominant}")
    else:
        print("Player 2 has no strictly dominant strategy.")

        
        
sample_matrix = [[(6, 4), (8, 1)], [(4, 6), (7, 3)]]
advertise_matrix = [[(10, 10), (15, 5)], [(5, 15), (7, 7)]]
prisoners_dilemma = [[(2, 2), (5, 0)], [(0, 5), (4, 4)]]
stag_hunt = [[(5, 5), (0, 3)], [(3, 0), (3, 3)]]

print("Testing Image Example:")
find_dominant_strategies(sample_matrix)

print("Testing (i) Advertise Matrix:")
find_dominant_strategies(advertise_matrix)

print("Testing (ii) Prisoner's Dilemma:")
find_dominant_strategies(prisoners_dilemma)

print("Testing (iii) Stag Hunt:")
find_dominant_strategies(stag_hunt)
```

```python
#Q3
def run_iesds(matrix):
    num_rows = len(matrix)
    num_cols = len(matrix[0])

    active_rows = list(range(num_rows))
    active_cols = list(range(num_cols))

    while True:
        eliminated_in_this_round = False

        row_to_eliminate = None
        for r1 in active_rows:
            for r2 in active_rows:
                if r1 == r2:
                    continue

                if all(
                    matrix[r1][c][0] < matrix[r2][c][0] for c in active_cols
                ):
                    row_to_eliminate = r1
                    print(f"Eliminated Row {r1} because it is strictly dominated by Row {r2}." )
                    break
            if row_to_eliminate is not None:
                break

        if row_to_eliminate is not None:
            active_rows.remove(row_to_eliminate)
            eliminated_in_this_round = True
            continue  

        col_to_eliminate = None
        for c1 in active_cols:
            for c2 in active_cols:
                if c1 == c2:
                    continue

                if all(
                    matrix[r][c1][1] < matrix[r][c2][1] for r in active_rows
                ):
                    col_to_eliminate = c1
                    print(f"Eliminated Column {c1} because it is strictly dominated by Column {c2}.")
                    break
            if col_to_eliminate is not None:
                break

        if col_to_eliminate is not None:
            active_cols.remove(col_to_eliminate)
            eliminated_in_this_round = True
            continue

        if not eliminated_in_this_round:
            break

    print("Final Results")
    print(f"Surviving Rows (Player 1): {active_rows}")
    print(f"Surviving Columns (Player 2): {active_cols}")

    # Check if a unique prediction outcome is found
    if len(active_rows) == 1 and len(active_cols) == 1:
        final_r = active_rows[0]
        final_c = active_cols[0]
        print(f"Predicted Outcome: Cell ({final_r}, {final_c}) with Payoffs {matrix[final_r][final_c]}")
    else:
        print("Multiple strategies survived. No unique predicted outcome.")


test_matrix = [
    [(4, 3), (5, 1), (6, 2), (1, 0)], 
    [(5, 6), (7, 2), (8, 3), (3, 4)], 
    [(2, 2), (3, 5), (4, 1), (2, 9)],  
]

run_iesds(test_matrix)
```

```python
#Q4
def find_nash_equilibria(matrix):
    num_rows = len(matrix)
    num_cols = len(matrix[0])

    p1_best_responses = set()
    p2_best_responses = set()

    for c in range(num_cols):
        max_p1_payoff = max(matrix[r][c][0] for r in range(num_rows))
        for r in range(num_rows):
            if matrix[r][c][0] == max_p1_payoff:
                p1_best_responses.add((r, c))

    for r in range(num_rows):
        max_p2_payoff = max(matrix[r][c][1] for c in range(num_cols))
        for c in range(num_cols):
            if matrix[r][c][1] == max_p2_payoff:
                p2_best_responses.add((r, c))

    nash_equilibria = p1_best_responses.intersection(p2_best_responses)

    print("Nash Equilibria Analysis")
    if nash_equilibria:
        print(f"Found {len(nash_equilibria)} Pure-Strategy Nash Equilibria:")
        for eq in sorted(nash_equilibria):
            r, c = eq
            print(
                f"  -> Cell ({r}, {c}) with Payoffs P1: {matrix[r][c][0]}, P2: {matrix[r][c][1]}"
            )
    else:
        print("No Pure-Strategy Nash Equilibrium found.")
    print("-" * 32 + "\n")


prisoners_dilemma = [[(2, 2), (5, 0)], [(0, 5), (4, 4)]]

stag_hunt = [[(5, 5), (0, 3)], [(3, 0), (3, 3)]]

chicken = [[(0, 0), (-1, 1)], [(1, -1), (-10, -10)]]

battle_of_the_sexes = [[(3, 2), (0, 0)], [(0, 0), (2, 3)]]

slide_matrix_3x3 = [
    [(2, 4), (1, 3), (0, 5)], 
    [(3, 1), (2, 2), (4, 3)],
    [(1, 2), (4, 4), (2, 1)], 

print("Testing (i) Prisoner's Dilemma:")
find_nash_equilibria(prisoners_dilemma)

print("Testing (ii) Stag Hunt:")
find_nash_equilibria(stag_hunt)

print("Testing (iii) Chicken:")
find_nash_equilibria(chicken)

print("Testing (iv) Battle of the Sexes:")
find_nash_equilibria(battle_of_the_sexes)

print("Testing (v) Slide 3x3 Matrix:")
find_nash_equilibria(slide_matrix_3x3)
```

```python
#Q5
import itertools


def find_n_player_nash(strategies, payoffs):
    num_players = len(strategies)
    all_profiles = list(itertools.product(*strategies))
    nash_equilibria = []

    for profile in all_profiles:
        is_nash = True

        for player_idx in range(num_players):
            current_payoff = payoffs[profile][player_idx]
            player_strategies = strategies[player_idx]

            for alternative_strategy in player_strategies:
                if alternative_strategy == profile[player_idx]:
                    continue

                mutated_profile = list(profile)
                mutated_profile[player_idx] = alternative_strategy
                mutated_profile = tuple(mutated_profile)

                deviation_payoff = payoffs[mutated_profile][player_idx]

                if deviation_payoff > current_payoff:
                    is_nash = False
                    break

            if not is_nash:
                break

        if is_nash:
            nash_equilibria.append(profile)

    print(f"{num_players}-Player Nash Equilibrium Analysis")
    if nash_equilibria:
        print(f"Found {len(nash_equilibria)} Pure-Strategy Nash Equilibria:")
        for eq in nash_equilibria:
            print(f"  -> Profile: {eq} | Payoffs: {payoffs[eq]}")
    else:
        print("No Pure-Strategy Nash Equilibrium found in this game setup.")
    print("-" * 40 + "\n")


firm_strategies = [["C", "D"], ["C", "D"], ["C", "D"]]
firm_payoffs = {
    ("C", "C", "C"): [4, 4, 4],
    ("C", "C", "D"): [1, 1, 6],
    ("C", "D", "C"): [1, 6, 1],
    ("C", "D", "D"): [0, 3, 3],
    ("D", "C", "C"): [6, 1, 1],
    ("D", "C", "D"): [3, 0, 3],
    ("D", "D", "C"): [3, 3, 0],
    ("D", "D", "D"): [2, 2, 2],
}

find_n_player_nash(firm_strategies, firm_payoffs)
```
